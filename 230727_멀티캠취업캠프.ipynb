```python
prices = [7,1,5,3,6,4]

def max_profit(prices):
    min_price=prices[0]
    max_profit=0

    for price in prices:
        if price < min_price:
            min_price = price #min_price는 계속 1
        elif price - min_price > max_profit:
            max_profit = price - min_price

    return max_profit
```

```python
def max_profit(prices):
    max_profit = 0
    min_price = float('inf') #<--컴퓨터가 표현할 수 있는 최대값 #양의 무한대로 초기화

    for price in prices:
        min_price = min(min_price,price)
        max_profit = max(max_profit, price - min_price)
    return max_profit

prices=[7,1,5,3,6,4]
result=max_profit(prices)
print("최대 이익:",result)
```

```python
def stock_profit(price):
    buy = min(price)
    buy_index=price.index(buy)
    sell=max(price[buy_index+1:])
    profit=sell-buy
    return profit
```

```python
stock_profit(prices)
```

```python
prices=[7,1,5,3,6,4]
def profit(prices):
    low = prices.index(min(prices))
    updated=prices[low:]
    print("최대이익은", max(updated)-min(prices))
```

```python
def read2d(file1):
    with open(file1,'r') as f:
        m=[i.split(' ') for i in f]
        list1=[]
        for line in m:
            for i,j in enumerate(line):
                if i>=len(list1):
                    list1.append([])
                list1[i].append(j)
        m = [list(map(int, num)) for num in list1]
        #print(m)
        return m

read2d('data.txt')
```

##5. 클래스

<파이썬 3 바이블, 이강성>

- 클래스는 새로운 이름 공간을 지원하는 단위

- 이 이름 공간에는 함수와 변수가 포함될 수 있다.

- 클래스를 정의하는 것은 새로운 자료형을 하나 만드는 것이다.

- 인스턴스는 이 자료형의 객체를 생성하는 것이다.

- 클래스 인스턴스 : 클래스를 호출하여 생성된 객체

- 클래스 인스턴스는 클래스의 실제 객체

- 클래스 인스턴스 객체 : 클래스 인스턴스와 같은 의미. 인스턴스 객체라 부르기도

- 멤버 : 클래스 혹은 클래스 인스턴스 공간에 정의된 변수

- 메서드 : 클래스 공간에 정의된 함수

- 속성(Attribute) : 멤버와 메서드 전체를 가리킴
