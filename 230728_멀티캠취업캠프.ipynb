```python
class FourCal:
    def __init__(self, first, second):
        self.setdata(first, second)

    def setdata(self, first, second):
        self.first = first
        self.second = second #다시 instanvce variale을 넣을 때도 있기에 setdata 한 것임

    def add(self):
        return self.first + self.second

    def mul(self):
        return self.first * self.second

    def sub(self):
        return self.first - self.second

    def div(self):
        return self.first / self.second
```

```python
class SafeFourCal(FourCal):   #상속 #상속관계가 되면 이걸 먼저 씀
    def div(self):
        if self.second == 0:
            return 0
        return self.first/self.second
```

```python
class FourCal:
    def __init__(self, first, second):
        self.setdata(first, second)

    def setdata(self, first, second):
        self.first = first
        self.second = second #다시 instanvce variale을 넣을 때도 있기에 setdata 한 것임

    def add(self):
        return self.first + self.second

    def mul(self):
        return self.first * self.second

    def sub(self):
        return self.first - self.second

    def div(self):
        if self.second == 0:
            raise ValueError("두 번째 값은 0이 아니어야 합니다.")
        return self.first / self.second
```

```python
# 클래스 인스턴스 생성
calc = FourCal(10,2)

# 각 메서드 호출하여 결과 출력
print(calc.add())  # 출력: 30 (10 + 20)
print(calc.mul())  # 출력: 200 (10 * 20)
print(calc.sub())  # 출력: -10 (10 - 20)
print(calc.div())  # 출력: 0.5 (10 / 20)
```
